## **Section 5.4: Time it Takes to Generate Attack Public Keys**

- We compute how long it takes an adversary with a certain number of HSDirs in the network to obtain a set of 900 v3 attack public keys in section 5.4. We repeat this computation ten times and present the results in figure 5.
- Below, we show how we obtain figure 5 from the simulation data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from os.path import join, abspath
```

```python
DATA_DIR = abspath("./3_data_generate-attack-public-keys")
```

```python
df = pd.read_csv(join(DATA_DIR, "times_atk_pubkeys_gen.csv"))
```

```python
df.groupby(["hsdir_matches_at_least", "pubkeys_list_size", "num_adv_hsdirs"]).time.describe()
```

```python
df_atl_1 = df.loc[df["hsdir_matches_at_least"] == 1].copy()
df_atl_2 = df.loc[df["hsdir_matches_at_least"] == 2].copy()
```

```python
df_atl_1.groupby(["hsdir_matches_at_least", "pubkeys_list_size", "num_adv_hsdirs"]).time.describe()[["count", "min", "50%", "mean", "max"]]
```

```python
df_atl_2.groupby(["hsdir_matches_at_least", "pubkeys_list_size", "num_adv_hsdirs"]).time.describe()[["count", "min", "50%", "mean", "max"]]
```

```python
# Attack public key generations so that the public keys map to at least 1 adversarial HSDir.

plt.figure(figsize=(12, 8))

sns.set()
sns.set(style="whitegrid", font_scale=1.33)
sns.set_style({"font.family": "Fira Sans Condensed"})

fig = sns.pointplot(data=df_atl_1, x="pubkeys_list_size", y="time", hue="num_adv_hsdirs")

leg = fig.axes.legend(loc="upper left", title="Adversarial HSDirs")

fig.set(ylim=(0.0, 11.0), yticks=np.arange(0.0, 11.1, 1.0))

plt.xlabel("Number of attack public keys")
plt.ylabel("Time to generate (s)")

plt.show()
```

```python
df_atl_2["time_min"] = df_atl_2["time"] / 60.0
df_atl_2.head(20)
```

```python
df_atl_2.groupby(["hsdir_matches_at_least", "pubkeys_list_size", "num_adv_hsdirs"]).time_min.describe()[["count", "min", "50%", "mean", "max"]]
```

```python
# Attack public key generations so that the public keys map to at least 2 adversarial HSDirs.

plt.figure(figsize=(12, 8))

sns.set()
sns.set(style="whitegrid", font_scale=1.75)
sns.set_style({"font.family": "Fira Sans Condensed"})

# The default confidence interval size for pointplots is 95.
# Means: The error bars (ending in the caps) comprise the 95% CI of the data.
fig = sns.pointplot(
    data = df_atl_2, x = "pubkeys_list_size",
    y = "time_min", hue = "num_adv_hsdirs",
    capsize = 0.025)

leg = fig.axes.legend(loc = "upper left", title = "Adversarial HSDirs")

fig.set(ylim = (0.0, 65.0), yticks = np.arange(0.0, 65.1, 5.0))

plt.xlabel("Number of attack onion service public keys")
plt.ylabel("Time to generate (m)")

plt.show()
```
